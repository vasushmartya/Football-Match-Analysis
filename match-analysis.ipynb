# Shots Analysis

```python
shots = events[events['type'] == 'Shot']
shots.head()
```

```python
shots['shot_outcome']
```

```python
x = shots['shot_outcome'].unique()
heights = [len(shots[shots['shot_outcome'] == outcome]) for outcome in x]
heights
```

```python
import matplotlib.pyplot as plt
plt.style.use('dark_background')
plt.bar(x= x, height=heights, color='c')
plt.title("Shot Outcomes")
plt.ylabel("Number of shots")
plt.show()
```

```python
heights[0]/sum(heights)
```

## Inference

- Most shots were actually saved by the goalkeepers (9)
- Raw **Probability of a goal** being scored is `{0.1538}` i.e. 15.38%

- Now let us see the team wise distribution

```python
shot_barca = shots[shots['team'] == 'Barcelona']

x_barca = shot_barca['shot_outcome'].unique()
heights_barca = [len(shot_barca[shot_barca['shot_outcome'] == outcome]) for outcome in x_barca]
heights_barca
```

```python
shot_madrid = shots[shots['team'] == 'Real Madrid']

x_madrid = shot_madrid['shot_outcome'].unique()
heights_madrid = [len(shot_madrid[shot_madrid['shot_outcome'] == outcome]) for outcome in x_madrid]
heights_madrid
```

```python
plt.style.use('dark_background')
plt.bar(x= x_barca, height=heights_barca, color='#A50044')
plt.show()
```

```python
plt.bar(x= x_madrid, height=heights_madrid, color='w')
plt.show()
```

```python
barca_counts = shot_barca['shot_outcome'].value_counts()
madrid_counts = shot_madrid['shot_outcome'].value_counts()
import pandas as pd

df = pd.DataFrame({
    'Barcelona': barca_counts,
    'Real Madrid': madrid_counts
}).fillna(0)
df.plot(kind='bar', color=['#A50044', '#FFFFFF'])

plt.title("Shot Outcomes: Barcelona vs Real Madrid")
plt.xlabel("Shot Outcome")
plt.ylabel("Number of Shots")
plt.xticks(rotation=45)

plt.show()
```

```python
barca_goal_prob = shot_barca['shot_outcome'].value_counts(normalize=True)['Goal']

print(f"Probability of Barca shot -> goal: {barca_goal_prob:.3f} i.e. {barca_goal_prob:.2%}")

madrid_goal_prob = shot_madrid['shot_outcome'].value_counts(normalize=True)['Goal']

print(f"Probability of Madrid shot -> goal: {madrid_goal_prob:.3f} i.e. {madrid_goal_prob:.2%}")
```

```python
barca_save_prob = shot_madrid['shot_outcome'].value_counts(normalize=True)['Saved']

print(f"Probability of Madrid shot -> saved by barca: {barca_save_prob:.3f} i.e. {barca_save_prob:.2%}")

madrid_save_prob = shot_barca['shot_outcome'].value_counts(normalize=True)['Saved']

print(f"Probability of Barca shot -> saved by barca: {madrid_save_prob:.3f} i.e. {madrid_save_prob:.2%}")
```

## Inference

- Probability of a Madrid shot being converted into a goal is higher (18.75%) than Barcelona(10%), which was clear in the scoreline of 1-3
- Probability of Madrid Shot being saved by Barcelona Goalkeeper is higher (37.5%) than Real Madrid(30%), but since Real Madrid just had higher number of shots (16 vs 10), they scored more.
- Barcelona Defence line blocked more shots than Real Madrid Defence.

```python
from mplsoccer import Pitch

shots[['x','y']] = pd.DataFrame(
    shots['location'].tolist(),
    index=shots.index
)
pitch = Pitch(pitch_type='statsbomb', pitch_color='green', line_color='white',
              stripe=True, stripe_color='lime')

fig, ax = pitch.draw(figsize=(10,7))

goals = shots[shots['shot_outcome'] == "Goal"]
non_goals = shots[shots['shot_outcome'] != "Goal"]

pitch.scatter(
    non_goals['x'],
    non_goals['y'],
    ax=ax,
    color=non_goals['team'].apply(lambda x: '#A50044' if x == "Barcelona" else 'white'),
    marker='o',
    s = 50
)

pitch.scatter(
    goals['x'],
    goals['y'],
    ax=ax,
    c=goals['team'].apply(lambda x: '#A50044' if x == "Barcelona" else 'white'),
    marker='football',
    s = 100
)
from matplotlib.lines import Line2D

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Barcelona',
           markerfacecolor='#A50044', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Real Madrid',
           markerfacecolor='w', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Goal',
           markerfacecolor='gray', markersize=8),
]

ax.legend(handles=legend_elements, loc='upper left')
plt.title("Barcelona vs Madrid Shot Map")
plt.show()
```

## Inferene

- Barcelona had a huge amount of passes in the midfield, usually the pass going sideways as well in the midfield
- Madrid on the other hand had higher amount of passes in the backline, but as they moved forward their passing density shifted towards the wings, specially the left wing, where Vinicius Junior Plays

```python
barca_events = events[
    events['team'] == 'Barcelona'
]

barca_events = barca_events.dropna(subset=['location'])
barca_events[['x','y']] = pd.DataFrame(
    barca_events['location'].tolist(),
    index=barca_events.index
)

pitch = Pitch(
    pitch_type='statsbomb',
    pitch_color='green',
    line_color='white',
    stripe=True,
    stripe_color='lime'
)

fig, ax = pitch.draw(figsize=(10,7))
pitch.kdeplot(
    barca_events['x'],
    barca_events['y'],
    ax=ax,
    fill=True,
    levels=15,
    thresh=0.05,
    color='r',
    alpha = 0.8
)
plt.title(f"Barcelona Heatmap")
plt.show()
```

## Barcelona Inference

- Team seems to be very active on the left side, specially left-mid position

```python
madrid_events = events[
    events['team'] == 'Real Madrid'
]

madrid_events = madrid_events.dropna(subset=['location'])
madrid_events[['x','y']] = pd.DataFrame(
    madrid_events['location'].tolist(),
    index=madrid_events.index
)

pitch = Pitch(
    pitch_type='statsbomb',
    pitch_color='green',
    line_color='white',
    stripe=True,
    stripe_color='lime'
)

fig, ax = pitch.draw(figsize=(10,7))
pitch.kdeplot(
    madrid_events['x'],
    madrid_events['y'],
    ax=ax,
    fill=True,
    levels=15,
    thresh=0.05,
    color='r',
    alpha = 0.8
)
plt.title(f"Madrid Heatmap")
plt.show()
```

## Real Madrid Inference

# Defence Analysis

```python
def_events = events[events['type'].isin([
    'Pressure',
    'Interception',
    'Block',
    'Clearance',
    'Duel',
    'Ball Recovery'
])]
```

```python
team_def_actions = def_events['team'].value_counts()
team_def_actions
```

- Real Madrid had more defensive actions than Barcelona

```python
player_def_actions = def_events['player'].value_counts(ascending=False)
player_def_actions.head(10)
```

- Toni Kroos Did the most defensive work in the match.

```python
def_events = def_events.dropna(subset=['location'])

def_events[['x','y']] = pd.DataFrame(
    def_events['location'].tolist(),
    index=def_events.index
)
```

```python
barca_def = def_events[def_events['team'] == "Barcelona"]
madrid_def = def_events[def_events['team'] == "Real Madrid"]
fig, ax = pitch.draw(figsize=(10,7))
pitch.kdeplot(
    barca_def['x'],
    barca_def['y'],
    ax=ax,
    fill=True,
    alpha = 0.8,
    color = 'r'
)
plt.title("Barcelona Defensive Heatmap")
plt.show()
fig, ax = pitch.draw(figsize=(10,7))
pitch.kdeplot(
    madrid_def['x'],
    madrid_def['y'],
    ax=ax,
    fill=True,
    alpha = 0.8,
    color = 'b'
)
plt.title("Madrid Defensive Heatmap")
plt.show()
```

## Inference

- It seems like Real Madrid handled the threat of Lionel Messi really well
- And that was more than enough for Real Madrid to stop the Barcelona Attacks and not let them score too many goals
- The only goal Barcelona scored was from Ansu Fati, who wasn't well defended according to the defensive heatmap
- Out of the 3 goals, 2 were non-penalty goals which were both scored by players who play on the right side of the pitch :- Valverde and Luka Modric

```python
fig, ax = pitch.draw(figsize=(10,7))
pitch.scatter(
    def_events[def_events['team'] == 'Barcelona']['x'],
    def_events[def_events['team'] == 'Barcelona']['y'],
    ax=ax,
    color= '#A50044'
)
plt.title("Barcelona Defence events")
plt.show()
```

```python
fig, ax = pitch.draw(figsize=(10,7))
pitch.scatter(
    def_events[def_events['team'] == 'Real Madrid']['x'],
    def_events[def_events['team'] == 'Real Madrid']['y'],
    ax=ax,
    color= 'w'
)
plt.title("Madrid Defence events")
plt.show()
```

```python
passes_by_team = passes['team'].value_counts()
def_actions_by_team = def_events['team'].value_counts()

teams = passes_by_team.index.tolist()

team1 = teams[0]
team2 = teams[1]

ppda_team1 = passes_by_team[team2] / def_actions_by_team[team1]
ppda_team2 = passes_by_team[team1] / def_actions_by_team[team2]

print(f"{team1} PPDA: {ppda_team1:.2f}")
```

```python

teams = [team1, team2]
ppda = [ppda_team1, ppda_team2]

plt.bar(teams, ppda, color='c')

plt.title("Pressing Intensity (PPDA)")
plt.ylabel("Passes per Defensive Action")

plt.show()
```

- Real Madrid seems to be pressing more efficiently and hardly than Barcelona

# Final Summary

In this project, match event data from the StatsBomb dataset was analyzed to study different aspects of team performance in a football match between Barcelona and Real Madrid. The analysis focused on several key areas including shot creation, passing patterns, player positioning, defensive actions, pressing intensity, and goalkeeper performance.

Shot analysis revealed differences in attacking efficiency and the spatial distribution of chances. Passing analysis helped evaluate ball progression and team build-up strategies, while player heatmaps illustrated how players occupied different zones of the pitch. Defensive events and the PPDA metric provided insight into pressing intensity and defensive behavior. Finally, goalkeeper actions such as saves and goals conceded were examined to evaluate shot-stopping performance.

Overall, the project demonstrates how event-level football data can be used to perform detailed tactical and statistical analysis of a match. Using Python tools such as `statsbombpy`, `pandas`, `matplotlib`, and `mplsoccer`, complex match dynamics can be explored through data-driven visualizations and metrics.
